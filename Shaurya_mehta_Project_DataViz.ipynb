```python
# Importing the libraries
import pandas as pd
import seaborn as sns
import csv 
import statistics
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the data from csv file
dataset = pd.read_csv('your path/D24_MAB_FY19P1to7.csv')
```

```python
# Observing initial observations of the dataset to gain a basic data understanding
dataset.head(10)
```

```python
# Get shape of the dataset
row, col = dataset.shape
print('Number of Dimensions:', dataset.ndim ,', having', row, "rows and", col, 'columns' )
```

```python
# Renaming the column headers
dataset.columns = ['WarehouseNo', 'RegisterNo','TransactionNo','FiscalPeriod','DateOfSale', 'TransactionTime', 'ItemNo','DailySalesUnites', 'DailySalesSell','Resell','Refund', 'DeptNo','CatCode1','CatCode2','CatCode3','ItemDes1','ItemDes2', 'RegionCode', 'District', 'City', 'State','MembershipIPK','AccountType','ExeFlag']
```

```python
dataset.head(5)
```

```python
#datatypes of the columns of the dataset
dataset.dtypes
```

```python
# Changing the datatype of DateOfSale from object to datetime
dataset['DateOfSale'] = pd.to_datetime(dataset['DateOfSale'])
```

```python
# Checking for null values
null_counts = dataset.isnull().sum()
null_counts
```

```python
# Replacing null values by N
dataset["Resell"].fillna("N", inplace = True)
dataset["Refund"].fillna("N", inplace = True)
dataset["Refund"].fillna("N", inplace = True)
```

```python
# Replacing null values by zero
dataset = dataset.fillna(0)
```

```python
# Deleting all the transactions having TransactionNo = zero
dataset.drop(dataset[dataset['TransactionNo'] == 0].index , inplace=True)
```

```python
# Changing the datatypes
dataset['WarehouseNo'] = dataset['WarehouseNo'].astype('int')
dataset['RegisterNo'] = dataset['RegisterNo'].astype('int')
dataset['TransactionNo'] = dataset['TransactionNo'].astype('int')
dataset['FiscalPeriod'] = dataset['FiscalPeriod'].astype('int')
dataset['TransactionTime'] = dataset['TransactionTime'].astype('int')
dataset['AccountType'] = dataset['AccountType'].astype('int')
dataset['ItemNo'] = dataset['ItemNo'].astype('int')
```

```python
# Get shape of the dataset
row, col = dataset.shape
print('Number of Dimensions:', dataset.ndim ,', having', row, "rows and", col, 'columns' )
```
